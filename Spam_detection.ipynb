```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to lemmatize and lowercase text
def lemmatize_and_lowercase(text):
    stop_words = set(stopwords.words('english'))
    words = text.split()

    # Remove stopwords
    filtered_words = []
    for word in words:
        if word not in stop_words:
            filtered_words.append(word)

    # Lemmatize and lowercase
    lemmatized_words = []
    for word in filtered_words:
        lemmatized_words.append(lemmatizer.lemmatize(word).lower())

    # Join the lemmatized and lowercase words
    result = ' '.join(lemmatized_words)
    return result

# Apply the function to the 'Message' column in your DataFrame
df['Message'] = df['Message'].apply(lemmatize_and_lowercase)

# Display the updated DataFrame
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Category']= le.fit_transform(df['Category'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000)

x = tfidf.fit_transform(df['Message']).toarray()
```

```python
y=df['Category']
```

```python
y
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.neural_network import MLPClassifier

mlp_classifier_model = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=1000)
mlp_classifier_model.fit(x_train, y_train)

prediction = mlp_classifier_model.predict(x_test)

print("Accuracy score: " + str(accuracy_score(y_test, prediction)))
```
